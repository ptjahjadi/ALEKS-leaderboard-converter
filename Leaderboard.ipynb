```python
# ALEKS Leaderboard Creator
# Python Code by Patrick Tjahjadi

# ----- Imported Libraries -----
# Pandas: required for DataFrame implementation and DateTime conversion
import pandas as pd
from pandas import DataFrame as df
# -------------------------------
```

```python
# Read the Leaderboard CSV
leaderboard = pd.read_csv('sample_report.csv', encoding='ISO-8859-1', skiprows = 10)
```

```python
# Rename the first column into student name
leaderboard = leaderboard.rename(columns={'Unnamed: 0': 'Student Name'})

# Replace unknown values '-' to 0 for numerical computation
leaderboard = leaderboard.replace('-', '0')
```

```python
# Convert the time format of hours and minutes into minutes.
# Assume that each weekly dataset has 9 instances of time columns.
for i in range(len(leaderboard)):
    try:
        if "h" in leaderboard["Time"][i]:
            time = pd.to_datetime(leaderboard["Time"][i], format = '%Hh%Mm')
            leaderboard["Time"][i] = time.hour * 60 + time.minute
        else:
            time = pd.to_datetime(leaderboard["Time"][i], format = '%Mm')
            leaderboard["Time"][i] = time.minute
    except TypeError:
        pass
    except ValueError:
        pass
    try:
        if "h" in leaderboard["Time.1"][i]:
            time = pd.to_datetime(leaderboard["Time.1"][i], format = '%Hh%Mm')
            leaderboard["Time.1"][i] = time.hour * 60 + time.minute
        else:
            time = pd.to_datetime(leaderboard["Time.1"][i], format = '%Mm')
            leaderboard["Time.1"][i] = time.minute
    except TypeError:
        pass
    except ValueError:
        pass
    try:
        if "h" in leaderboard["Time.2"][i]:
            time = pd.to_datetime(leaderboard["Time.2"][i], format = '%Hh%Mm')
            leaderboard["Time.2"][i] = time.hour * 60 + time.minute
        else:
            time = pd.to_datetime(leaderboard["Time.2"][i], format = '%Mm')
            leaderboard["Time.2"][i] = time.minute
    except TypeError:
        pass
    except ValueError:
        pass
    try:
        if "h" in leaderboard["Time.3"][i]:
            time = pd.to_datetime(leaderboard["Time.3"][i], format = '%Hh%Mm')
            leaderboard["Time.3"][i] = time.hour * 60 + time.minute
        else:
            time = pd.to_datetime(leaderboard["Time.3"][i], format = '%Mm')
            leaderboard["Time.3"][i] = time.minute
    except TypeError:
        pass
    except ValueError:
        pass
    try:
        if "h" in leaderboard["Time.4"][i]:
            time = pd.to_datetime(leaderboard["Time.4"][i], format = '%Hh%Mm')
            leaderboard["Time.4"][i] = time.hour * 60 + time.minute
        else:
            time = pd.to_datetime(leaderboard["Time.4"][i], format = '%Mm')
            leaderboard["Time.4"][i] = time.minute
    except TypeError:
        pass
    except ValueError:
        pass
    try:
        if "h" in leaderboard["Time.5"][i]:
            time = pd.to_datetime(leaderboard["Time.5"][i], format = '%Hh%Mm')
            leaderboard["Time.5"][i] = time.hour * 60 + time.minute
        else:
            time = pd.to_datetime(leaderboard["Time.5"][i], format = '%Mm')
            leaderboard["Time.5"][i] = time.minute
    except TypeError:
        pass
    except ValueError:
        pass
    try:
        if "h" in leaderboard["Time.6"][i]:
            time = pd.to_datetime(leaderboard["Time.6"][i], format = '%Hh%Mm')
            leaderboard["Time.6"][i] = time.hour * 60 + time.minute
        else:
            time = pd.to_datetime(leaderboard["Time.6"][i], format = '%Mm')
            leaderboard["Time.6"][i] = time.minute
    except TypeError:
        pass
    except ValueError:
        pass
    try:
        if "h" in leaderboard["Time.7"][i]:
            time = pd.to_datetime(leaderboard["Time.7"][i], format = '%Hh%Mm')
            leaderboard["Time.7"][i] = time.hour * 60 + time.minute
        else:
            time = pd.to_datetime(leaderboard["Time.7"][i], format = '%Mm')
            leaderboard["Time.7"][i] = time.minute
    except TypeError:
        pass
    except ValueError:
        pass
    try:
        if "h" in leaderboard["Time.8"][i]:
            time = pd.to_datetime(leaderboard["Time.8"][i], format = '%Hh%Mm')
            leaderboard["Time.8"][i] = time.hour * 60 + time.minute
        else:
            time = pd.to_datetime(leaderboard["Time.8"][i], format = '%Mm')
            leaderboard["Time.8"][i] = time.minute
    except TypeError:
        pass
    except ValueError:
        pass
```

```python
# Initialise the Python dictionaries from time, learned and learned per hour
time_dict = {}
learned_dict = {}
learned_per_hour_dict = {}

for i in range(len(leaderboard)):
    time_dict[leaderboard["Student Name"][i]] = int(float(leaderboard["Time.7"][i]))
    learned_dict[leaderboard["Student Name"][i]] = int(float(leaderboard["Learned.7"][i]))
    learned_per_hour_dict[leaderboard["Student Name"][i]] = int(float(leaderboard["Learned/Hour.7"][i]))
```

```python
# Sort the dictionaries for ranking purposes

sorted_time_dict = sorted(time_dict.items(), key=lambda x: x[1], reverse = True)
sorted_learned_dict = sorted(learned_dict.items(), key=lambda x: x[1], reverse = True)
sorted_lph_dict = sorted(learned_per_hour_dict.items(), key=lambda x: x[1], reverse = True)
```

```python
# Create the time leaderboard in a separate text file
text_file = open("time_leaderboard.txt", "w")

def write_to_file():
    global text_file
    text_file.write("Position")
    text_file.write("          ")
    text_file.write("Name")
    text_file.write("                          ")
    text_file.write("Time (Minutes)")
    text_file.write("\n")
    for i in range(len(sorted_time_dict)):
        text_file.write(str(i+1).ljust(15))
        name = str(sorted_time_dict[i][0])
        text_file.write(name.ljust(30))
        text_file.write("     ")
        text_file.write(str(sorted_time_dict[i][1]).ljust(20))
        text_file.write("\n")
    text_file.close()

write_to_file()
```
